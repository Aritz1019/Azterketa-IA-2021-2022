```python
import pandas as pd
pd.set_option('display.max_columns', 500)
```

## 1- Kargatu Fitxategia (1PT)

```python
datatrain = pd.read_csv('FifaTrain.csv')
datatrain.head()
```

## 2- AurreProzesaketa (3 PT)
### Prediktoreak ez diren aldagaiak kendu (0.5pt)


```python
# Undamed sobra
datatrain.drop(['Unnamed: 0'], inplace=True, axis=1)
# Id no da ninguna informacion
datatrain.drop(['Name'], inplace=True, axis=1)
# El nomrbre no influye en el potencial del jugador
datatrain.drop(['ID'], inplace=True, axis=1)
# La foto da igual
datatrain.drop(['Photo'], inplace=True, axis=1)
# En el caso del potencial la nacionalidad da igual, si fuese el UT si tendria que ver
datatrain.drop(['Nationality'], inplace=True, axis=1)
# La bandera tambien da igual
datatrain.drop(['Flag'], inplace=True, axis=1)
# Club logo fuera
datatrain.drop(['Club Logo'], inplace=True, axis=1)
# Que tan buena reputacion tiene no tiene relacion
datatrain.drop(['International Reputation'], inplace=True, axis=1)
# Que en el juego tenga la cara real no influye
datatrain.drop(['Real Face'], inplace=True, axis=1)
# El dorsal tampoco influye
datatrain.drop(['Jersey Number'], inplace=True, axis=1)
# Cuando se unio al club da igual
datatrain.drop(['Joined'], inplace=True, axis=1)


datatrain.head()
```

### NaN-ak kentzeko estrategia garatu (1 pt)
Estrategia defenda zenezakete, deskripzio batekin


```python
pd.set_option('display.max_rows', None)
total = datatrain.isnull().sum().sort_values(ascending = False)
percent = round(datatrain.isnull().sum().sort_values(ascending = False)/len(datatrain)*100, 2)
temp = pd.concat([total, percent], axis = 1,keys= ['Total', 'Percent'])
temp.loc[(temp['Total'] > 0)]
```

```python
numericos = datatrain.select_dtypes(include=['int64', 'float64'])
categoricos = datatrain.select_dtypes(include=['object'])
print(numericos.columns)
print(categoricos.columns)
```

```python
numericos_nan = numericos.isna().columns
categoricos_nan = categoricos.isna().columns

# el porcentaje de NaN no es muy alto asique para probar se va a usar la mediana

for columna in numericos_nan:
    numericos[columna].fillna(numericos[columna].median(), inplace=True)

# en el caso de los categoricos lo mas comun es que una persona sea diestra 
# aparte de eso se pondra que trabaja en la media tanto en ataque como en defensa
categoricos['Preferred Foot'].fillna('Right', inplace=True)
categoricos['WorkFront'].fillna('Medium', inplace=True)
categoricos['WorkBack'].fillna('Medium', inplace=True)

datatrain = pd.concat([numericos, categoricos], axis=1)

datatrain.head()
```

```python
#Ahora no tendria que haber ninguna celda con null
print('Number of missing values:')
datatrain.isnull().sum().sum()
```

### Aldagai kategorikoak landu (1pt)

```python
categoricos = datatrain.select_dtypes(include=['object'])
categoricos.head()
```

```python
# Preferred foot con one hot encoder ya que ninguna opcion es mejor que la otra
from sklearn.preprocessing import LabelEncoder, OneHotEncoder

foot_column = datatrain[["Preferred Foot"]]

hot_encoder = OneHotEncoder(sparse=False)
foot_encoded = hot_encoder.fit_transform(foot_column)

foot_encoded
```

```python
feature_names = hot_encoder.get_feature_names(input_features=["Preferred Foot"])
foot_encoded = pd.DataFrame(foot_encoded, columns=feature_names)
foot_encoded.drop(['Preferred Foot_Left'], inplace=True, axis=1)
foot_encoded.head()

datatrain = pd.concat([datatrain, foot_encoded], axis=1)
datatrain.drop(['Preferred Foot'], inplace=True, axis=1)
datatrain.head()
```

```python
# En el caso de workfront y workback si hay relacion entre las distintas opciones por eso mismo se va a usar un labelencoder

label_encoder = LabelEncoder()
datatrain['WorkFront'] = label_encoder.fit_transform(datatrain['WorkFront'])
datatrain['WorkBack'] = label_encoder.fit_transform(datatrain['WorkBack'])
```

```python
first_column = datatrain.pop('Potential')
datatrain.insert(0, 'Potential', first_column)
datatrain.head()
```

### Datuen Eskalatzea, Diskretizazioa, Aldagai Aukeraketa (agian ez dira guzti hauek behar, emaitza ona lortzeko)(0.5pt)
Hemen ere defenda dezakezua, deskripzio txiki batekin

```python
# La edad, el peso y la altura se va a discretizar, metiendolos en diferentes grupos
bins_edad=[0, 18, 22, 25, 28, 100]
bins_altura=[0, 5, 5.3, 5.6, 5.9, 8]
bins_peso=[0, 50, 70, 90, 110, 130]
labels=["1", "2", "3", "4", "5"]
datatrain['Age'] = pd.cut(datatrain['Age'], bins=bins_edad, labels=labels)
datatrain['Height'] = pd.cut(datatrain['Height'], bins=bins_altura, labels=labels)
datatrain['Weight'] = pd.cut(datatrain['Weight'], bins=bins_peso, labels=labels)

datatrain.head()
```

```python
# Vamos a normalizar las estadisticas de los jugadores
columnas = ['LS', 'ST', 'RS', 'LW', 'LF', 'CF', 'RF', 'RW',
       'LAM', 'CAM', 'RAM', 'LM', 'LCM', 'CM', 'RCM', 'RM', 'LWB', 'LDM',
       'CDM', 'RDM', 'RWB', 'LB', 'LCB', 'CB', 'RCB', 'RB', 'Crossing',
       'Finishing', 'HeadingAccuracy', 'ShortPassing', 'Volleys', 'Dribbling',
       'Curve', 'FKAccuracy', 'LongPassing', 'BallControl', 'Acceleration',
       'SprintSpeed', 'Agility', 'Reactions', 'Balance', 'ShotPower',
       'Jumping', 'Stamina', 'Strength', 'LongShots', 'Aggression',
       'Interceptions', 'Positioning', 'Vision', 'Penalties', 'Composure',
       'Marking', 'StandingTackle', 'SlidingTackle', 'GKDiving', 'GKHandling',
       'GKKicking', 'GKPositioning', 'GKReflexes']
```

```python
from sklearn.preprocessing import StandardScaler
standardScaler = StandardScaler()
datosEstandar = datatrain.copy()
datosEstandar[columnas] = standardScaler.fit_transform(datosEstandar[columnas])
datosEstandar.head()
# al final vamos a normalizar en vez de estandarizar
```

```python
from sklearn.preprocessing import MinMaxScaler
minMaxScaler = MinMaxScaler()
datatrain[columnas] = minMaxScaler.fit_transform(datatrain[columnas])
datatrain.head()
```

## 3-Modelaketa eta balidazioa (4pt)
### Datu Zatiketa eta balidazio estrategia (1pt)
hemen, nahi bada erabilitakoa deskriba zenezakete (LABUR), hartu duzuen erabakiaren oinarria zein den niri explikatzeko


```python
from sklearn.model_selection import train_test_split
import numpy as np

X = datatrain.iloc[:, 1:].values
y = datatrain.iloc[:, 0].values

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 0)
```

### Modelaketa bera (2pt)
Hemen ere erabilitako modeloaren explikazio TXIKI bat jarri zenezakete, zergatik aukeratu duzuen explikatzeko


```python
# De los diferentes algoritmos que hay se va a utilizar el Multiple Linear Regression ya que hay mas de una variable para predecir el potencial del jugador
from sklearn.linear_model import LinearRegression

regressor = LinearRegression()
regressor.fit(X_train, y_train)

# Predicting the Test set results
y_pred = regressor.predict(X_test)
np.set_printoptions(precision=2)
print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1))
```

### Balidazioa (1pt) 
Overfitting-ik ba al duzu?

```python
from sklearn.model_selection import cross_val_score

cv = cross_val_score(regressor, X, y) #cv cuantos trozos quieres hacer
scores = pd.Series(cv)
scores
```

```python
from sklearn.metrics import r2_score, mean_squared_error

# print("r2 train: " + str(r2_score(y_train, y_train_pred)))
print("r2 test: " + str(r2_score(y_test, y_pred)))
print("rmse test: " + str(mean_squared_error(y_test, y_pred, squared=False)))
print("cross validation: " + str(scores.mean()))

# r2 eta cross validation gutxi gorabehera valio berdina daukate, beraz esan daiteke overfitting-a ez daukala
```
